# DHS Overlap

Peak set overlap with DHS Index peaks.

**Inputs**: 
1. Outputs of dhs_overlap.sh (dhs_overlap_out.txt). Typically counts fraction of peaks in peak set, which after resizing to center 500 bp have > 200 bp overlap with any peak in DHS Index.

```r
library(ggplot2)
```

```r
dhs_overlap = read.table("dhs_overlap_out.txt", sep='\t')
colnames(dhs_overlap) = c("Peak_set", "Frac", "Peak", "Total")
dhs_overlap$miss = 1-dhs_overlap$Frac
dhs_overlap
```

```r
options(repr.plot.width = 10, repr.plot.height = 8)

p = ggplot(dhs_overlap, aes(x=factor(Peak_set), y=100*miss)) +
geom_bar(stat="identity", color="white", fill="dodgerblue1") + 
xlab("Peak Set") +
ylab("% Peaks absent from DHS Index") +
ggtitle("Overlap with DHS Index") +
theme_minimal(base_size = 26) +
scale_y_continuous(expand = expansion(mult = c(0.01, .1))) +
theme(plot.title = element_text(hjust = 0.5),
      axis.ticks = element_line(colour = "black", size = .3),
     panel.border = element_rect(colour = "black", fill=NA, size=0.5),
     panel.grid.major.x = element_blank())

p
```

```r
ggsave("./subfigs/DHS_overlap.pdf", p, height=8, width=10, useDingbats=F)
```
