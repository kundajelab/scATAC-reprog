# Motifs export

Export list of consolidated motifs across states and assignments to modisco motifs per cell state.

```python
import os
```

```python
import pandas as pd
```

```python
cluster_config = pd.read_csv("../../figures_factory/configs/cluster.tsv", sep='\t')
cluster_config
```

### Consolidated motifs

Clustered with gimme motifs cluster.

```python
consolidated = pd.read_csv("/users/surag/kundajelab/scATAC-reprog/src/analysis/20211007_motif_consolidation/20210820_chrombpnet_lite/tfs_final_w_rep.tsv", 
           sep='\t', names=['gimme_cluster', 'Vierstra_match', 'rep_old', 'motif'])
consolidated.head()
```

```python
def get_new_name(x):
    # x: cts6_pattern_8
    pat = "_".join(x.split("_")[1:])
    c = cluster_config.iloc[int(x.split("_")[0][3:])-1]
    return "c" + str(c['new_cluster']) + "_" + c['description'].lower().replace('-', '_') + "_" + pat
```

```python
consolidated['representative'] = consolidated['rep_old'].transform(lambda x: get_new_name(x))
```

```python
consolidated
```

```python
consolidated[['motif', 'representative', 'Vierstra_match']].to_csv("/oak/stanford/groups/akundaje/surag/projects/reprog-package/chrombpnet/modisco/counts.consolidated.tsv",
                                                                  sep='\t', index=False)
```

### Write PFMs

Write trimmed PFMs corresponding to the representative motif.

```python
for _, x in consolidated.iterrows():
    pat = "_".join(x['rep_old'].split("_")[1:])
    clst = x['rep_old'].split("_")[0]
    os.symlink(f"/users/surag/kundajelab/scATAC-reprog/src/analysis/20211007_motif_consolidation/20210820_chrombpnet_lite/pfms/{clst}/{pat}.pfm",
              f"/oak/stanford/groups/akundaje/surag/projects/reprog-package/chrombpnet/modisco/consolidated_pfms/{x['motif']}.pfm")
```

### Per cluster assignments

Manually adjusted assignments. Done only for counts motifs.

```python
for _,x in cluster_config.iterrows():
    new_name = x['description'].lower().replace('-','_')
    os.symlink(f"/users/surag/kundajelab/scATAC-reprog/src/analysis/20230607_re_modisco_breakdown/manually_adjusted/cluster_idx{x['cluster']}.txt",
              f"/oak/stanford/groups/akundaje/surag/projects/reprog-package/chrombpnet/modisco/counts/c{x['new_cluster']}_{new_name}.counts.motifs.tsv")
```
