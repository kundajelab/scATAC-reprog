# Annotate Peaks with Other Peaks

Perform intersections of peak sets with other peaks sets.

```bash
BASE=~/kundajelab/scATAC-reprog/
ANALYSIS=$BASE/src/analysis/20200225_D2_bifurcation
RES=$BASE/resources/
```

```bash
# Intersect peaks in each cluster with H1-hESC ChIP peaks from GSE17917 (Lister et al 2009)

TOT=$(cat $RES/GSE17917/KLF4.hg38.bed $RES/GSE17917/SOX2.hg38.bed $RES/GSE17917/OCT4.hg38.bed | wc -l)
printf "TOTAL PEAKS : $TOT\n\n"

(printf "Cluster Intersect Total Percent\n" ;
for i in {1..10} 
 do INT=$(bedtools intersect -b <(cat $RES/GSE17917/KLF4.hg38.bed \
                                      $RES/GSE17917/SOX2.hg38.bed \
                                      $RES/GSE17917/OCT4.hg38.bed |  cut -f1-3) \
   -a $ANALYSIS/beds/20200227_heatmap_ecto_9_vs_repro_3_11_naive_n10/idx$i.bed -c \
   | awk '$4>0{s+=1} END{print s}') ; 
   TOT=$(cat  $ANALYSIS/beds/20200227_heatmap_ecto_9_vs_repro_3_11_naive_n10/idx$i.bed | wc -l ) ; 
   FRAC=$(echo "100*$INT/$TOT" | bc -l | awk '{printf "%.4f\n", $0}') ; 
   printf "$i $INT $TOT $FRAC\n" 
done)  | column -t
```
